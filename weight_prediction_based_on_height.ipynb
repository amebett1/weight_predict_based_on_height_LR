```python
import numpy as np 
import matplotlib.pyplot as plt
```

Data Visualization

```python
# Size (m2)
X = np.array([
    66.2070, 68.6208, 71.0346, 73.4484, 75.8622, 
    78.2760, 80.6898, 83.1036, 85.5174, 87.9312,
    90.3450, 92.7588, 95.1726, 97.5864, 100.0002, 
    102.4140, 104.8278, 107.2416, 109.6554, 112.0692
]).reshape(-1, 1)

# Price (mil VND)
y = np.array([
    1025.45, 1120.34, 1195.8, 1150.2, 1280.45, 
    1310.9, 1420.15, 1390.6, 1450.32, 1510.75,
    1480.2, 1595.4, 1640.85, 1610.1, 1750.5, 
    1820.3, 1780.95, 1890.4, 1950.25, 1910.8
]).reshape(-1, 1)
```

```python
plt.scatter(X,y)
plt.xlabel('Size (m2)')
plt.ylabel('Price (mil VND)')
plt.axis([50,150,1000,2000])
plt.grid()
plt.show()
```

Tìm Fitting Line

```python
new_column_2d = np.ones((X.shape[0], 1))
Xbar = np.concatenate((new_column_2d, X), axis=1)

# A và b
A = np.dot(Xbar.T, Xbar)
b = np.dot(Xbar.T, y)

print("--- Ma trận Xbar ---")
print(Xbar)
print("\n--- Ma trận A (Xbar.T * Xbar) ---")
print(A)
print("\n--- Vector b (Xbar.T * y) ---")
print(b)

w = np.dot(np.linalg.pinv(A),b) # A-dagger * b
print('\n w = ', w)
w0 = w[0][0]
w1 = w[1][0]
x0 = np.linspace(60 ,120) # Chiều cao
y0 = w1*x0 + w0 # Cân nặng
```

```python
# Fitting Line
plt.scatter(X,y)
plt.plot(x0,y0,color = 'red')
plt.grid()
plt.axis([50,150,1000,2000])
plt.show()
```

Prediction

```python
x1 = 155 # Thay số liệu
y1 = w1*x1 + w0
print('Price prediction of house where its area is %.2f (m2) is' %x1)
print(y1, 'million VND')
```
